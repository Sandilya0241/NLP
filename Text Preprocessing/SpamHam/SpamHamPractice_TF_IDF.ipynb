```python
import pandas as pd
messages = pd.read_csv("../Bag_Of_Words/SMSSpamCollection.txt", sep="\t", names=['label', 'message'])
messages
```

```python
import nltk
import re

nltk.download("stopwords")

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

corpus = []

for idx in range(len(messages)):
    message = messages['message'][idx]
    message = re.sub("[^a-zA-Z]", " ", str(message))
    words = message.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stopwords.words("english")]
    corpus.append(" ".join(words))

corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tdidfVectorizer = TfidfVectorizer(max_features=2500, ngram_range=(1,2),lowercase=True)

X = tdidfVectorizer.fit_transform(corpus).toarray()

X
```

```python
y = messages['label'].map({'ham':0,'spam':1}).values
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

X_train, X_text, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = MultinomialNB().fit(X_train, y_train)

y_pred=model.predict(X_text)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

ac=accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
cr=classification_report(y_test,y_pred)
print(ac)
print(cm)
print(cr)

```
