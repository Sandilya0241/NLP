```python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel, whose company specialized in building metal frameworks and structures."

'''
    What is Name Entity Recognition?
    It is the process of locating and classifying named entities in text into predefined categories such as the names of persons, organizations, locations, expressions of times, quantities, monetary values, percentages, etc.
    For example, in the above sentence, 
        "Eiffel Tower" is a location, 
        "Gustave Eiffel" is a person and 
        "1887 to 1889" is a time expression.
'''
```

```python
import nltk

nltk.download('maxent_ne_chunker_tab')
nltk.download('words')

words=nltk.word_tokenize(sentence)
tagged_elements = nltk.pos_tag(words)
named_entities = nltk.ne_chunk(tagged_elements)
nltk.ne_chunk(tagged_elements).draw()
print(named_entities)
```
